### [실습 1] 중심 경향 및 산포 측도 계산


아래 배열에서 평균, 중앙값, 최빈값, 분산, 표준편차, 범위를 계산하세요.


```python
	arr = np.array([10, 20, 30, 30, 40, 50, 60])
```

````python
import numpy as np

arr = np.array([10, 20, 30, 30, 40, 50, 60])

# 최빈값 구하기
values, counts = np.unique(arr, return_counts=True)
mode = values[np.argmax(counts)]

print("평균:", np.mean(arr))
print("중앙값:", np.median(arr))
print("최빈값:", mode)
print("분산:", np.var(arr))
print("표준편차:", np.std(arr))
print("범위:", np.max(arr) - np.min(arr))
````

### [실습 2] 정규분포와 균등분포의 난수 생성


정규분포와 균등분포에서 각각 난수를 생성하여, 평균과 표준편차를 계산하세요.

1. 평균 0, 표준편차 1인 정규분포 난수 1000개 생성
2. 0과 1 사이의 균등분포 난수 1000개 생성

````python
# 1. 평균 0, 표준편차 1인 정규분포 난수 1000개
normal_data = np.random.normal(
    loc=0, scale=1, size=1000
)

print("정규분포 평균:", np.mean(normal_data))
print("정규분포 표준편차:", np.std(normal_data))

# 2. 0 이상 1 미만인 균등분포 난수 1000개
uniform_data = np.random.uniform(
    low=0, high=1, size=1000
)

print("균등분포 평균:", np.mean(uniform_data))
print("균등분포 표준편차:", np.std(uniform_data))
````

### [실습 3] 평균 성적 구하기


아래와 같이 학생별 과목 성적 데이터가 주어졌을 때, 각 학생의 전체 평균 성적(수학, 영어, 과학)을 계산한 후, 이 평균값에 대한 분산과 표준편차를 구하세요.


```python
# 학생별 과목 성적 데이터
scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]
```

````python
import numpy as np

scores_math = [88, 92, 76, 85, 95, 92, 88, 90, 85, 88]
scores_english = [90, 85, 80, 87, 92, 89, 83, 91, 86, 88]
scores_science = [85, 88, 82, 89, 94, 86, 91, 88, 84, 87]

# 행: 학생, 열: 과목
scores = np.array([
    scores_math,
    scores_english,
    scores_science
]).T

# 학생별 과목 평균
student_means = np.mean(scores, axis=1)

print("학생별 평균:", student_means)
print("평균값의 분산:", np.var(student_means))
print("평균값의 표준편차:", np.std(student_means))
````
